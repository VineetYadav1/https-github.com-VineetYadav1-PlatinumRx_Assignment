```python
Given number of minutes, convert it into human readable form.
Example :
- 130 becomes “2 hrs 10 minutes”
- 110 becomes “1hr 50minutes”
```

```python
def minutes_to_hours_minutes(total_minutes):
    hours = total_minutes // 60
    remaining_minutes = total_minutes % 60
    return f"{hours} hrs {remaining_minutes} minutes"

print(minutes_to_hours_minutes(130))
print(minutes_to_hours_minutes(110))
print(minutes_to_hours_minutes(45))
```

```python
2) You are given a string, remove all the duplicates and print the unique string.Use loop in the python.
```

```python
def remove_duplicates(s):
    result = ""
    for char in s:
        if char not in result:
            result += char
    return result


print(remove_duplicates("programming"))
print(remove_duplicates("banana"))
print(remove_duplicates("aabbccdd"))
```
